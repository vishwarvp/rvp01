## Handle Outliers



```python
numerical_cols= taxis_data.select_dtypes(include=[ 'int64','float64']).columns

outliers_columns={}
for col in numerical_cols:
    Q1= taxis_data[col].quantile(0.25)
    Q3=taxis_data[col].quantile(0.75)
    IQR=Q3-Q1

    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR

    outliers=taxis_data[(taxis_data[col]<lower_limit) | (taxis_data[col]>upper_limit)]
    if len(outliers)>0:
        outliers_columns[col]=len(outliers)

print("Outliers_columns: ",outliers_columns)
```



## Boxplot

IQR Method → (Q1, Q3, IQR, Upper, Lower bounds)

```python
print("\nOutliers_Passengers")

# Quartiles
Q1 = taxis_data['passengers'].quantile(0.25)
Q2 = taxis_data['passengers'].quantile(0.50)
Q3 = taxis_data['passengers'].quantile(0.75)
IQR = Q3 - Q1

print("\nFirst Quartile:", Q1)
print("Second Quartile:", Q2)
print("Third Quartile:", Q3)
print("Minimum passengers:", taxis_data['passengers'].min())
print("Maximum passengers:", taxis_data['passengers'].max())
print("INTERQUARTILE RANGE:", IQR)

# Outlier limits
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Lower limit:", lower)
print("Upper limit:", upper)

# Detect outliers
outliers = taxis_data[(taxis_data['passengers'] < lower) | (taxis_data['passengers'] > upper)]
outliers.head()
```









## Feature Engineering

### Create new columns if required






```python
taxis_data['trip_duration_min'] = (taxis_data['dropoff'] - taxis_data['pickup']).dt.total_seconds() / 60
taxis_data.head(2)
```

```python
taxis_data['fare_per_km'] = taxis_data['fare'] / taxis_data['distance']
taxis_data.head(2)

```

```python
taxis_data['total_per_km'] = taxis_data['total'] / taxis_data['distance']

taxis_data.head(2)
```

### Transformations (Log, Standardization, Encoding)


```python
taxis_data[numerical_cols].skew()
```

```python
cols_to_log = ['distance', 'fare', 'tip', 'tolls','total']

for col in cols_to_log:
    taxis_data[col + '_log'] = np.log1p(taxis_data[col])

taxis_data
```

### Remove Irrelevant / Unnecessary Columns

#### Based on domain logic

```python

taxis_data = taxis_data.drop(['pickup_zone', 'dropoff_zone'], axis=1)
taxis_data
```

```python
taxis_data = taxis_data.drop(['pickup_borough','dropoff_borough'], axis=1)
taxis_data
```
