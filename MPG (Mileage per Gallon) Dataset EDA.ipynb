## Handle Outliers


```python
numerical_cols= mpg_df.select_dtypes(include=[ 'int64','float64']).columns

outliers_columns={}
for col in numerical_cols:
    Q1= mpg_df[col].quantile(0.25)
    Q3=mpg_df[col].quantile(0.75)
    IQR=Q3-Q1

    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR

    outliers=mpg_df[(mpg_df[col]<lower_limit) | (mpg_df[col]>upper_limit)]
    if len(outliers)>0:
        outliers_columns[col]=len(outliers)

print("Outliers_columns: ",outliers_columns)
```

## Boxplot

IQR Method → (Q1, Q3, IQR, Upper, Lower bounds)

```python
print("\nOutliers_horsepower")

# Quartiles
Q1 = mpg_df['horsepower'].quantile(0.25)
Q2 = mpg_df['horsepower'].quantile(0.50)
Q3 = mpg_df['horsepower'].quantile(0.75)
IQR = Q3 - Q1

print("\nFirst Quartile:", Q1)
print("Second Quartile:", Q2)
print("Third Quartile:", Q3)
print("Minimum horsepower:", mpg_df['horsepower'].min())
print("Maximum horsepower:", mpg_df['horsepower'].max())
print("INTERQUARTILE RANGE:", IQR)

# Outlier limits
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Lower limit:", lower)
print("Upper limit:", upper)

# Detect outliers
outliers = mpg_df[(mpg_df['horsepower'] < lower) | (mpg_df['horsepower'] > upper)]
outliers.head()
```

```python
plt.figure(figsize=(6,4))

plt.boxplot(
    mpg_df['horsepower'],
    patch_artist=True,        # enables fill color
    boxprops=dict(facecolor='lightblue', color='blue', linewidth=1.5),
    medianprops=dict(color='red', linewidth=2),
    whiskerprops=dict(color='darkblue', linewidth=1.5),
    capprops=dict(color='blue', linewidth=1.5),
    flierprops=dict(marker='o', markerfacecolor='orange', markersize=8, markeredgecolor='red')
)

plt.title("Boxplot of horsepower", fontsize=14)
plt.ylabel("Number of horsepower", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

### Numerical → Histogram

```python
num_cols = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']

for col in num_cols:
    plt.figure(figsize=(10, 6))

    # histogram with label (needed for legend)
    counts, bins, patches = plt.hist(mpg_df[col].dropna(), bins=30, label=col)

    # add random color for each bar
    for patch in patches:
        patch.set_facecolor(np.random.rand(3,))

    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.title(f"Distribution of {col}")
    plt.grid(True)
    plt.legend()       
    plt.show()
```

## Feature Engineering

### Create new columns if required


```python
#Horsepower per Liter (Engine Efficiency)
mpg_df['hp_per_litre'] = mpg_df['horsepower'] / mpg_df['displacement']
mpg_df
```

```python
#Acceleration Efficiency
mpg_df['power_accel_ratio'] = mpg_df['horsepower'] / mpg_df['acceleration']
```

```python
#engine size
mpg_df['engine_size']= pd.cut(mpg_df['displacement'],bins=[0,120,300,500],labels=['small','medium','large'])
mpg_df
```

```python
mpg_df['hp_category'] = pd.cut(mpg_df['horsepower'],
                               bins=[0, 100, 180, 300],
                               labels=['Low', 'Medium', 'High'])
```

### Transformations (Log, Standardization, Encoding)


```python
mpg_df[numerical_cols].skew()
```

```python
cols_to_log = ['mpg', 'displacement', 'horsepower', 'acceleration']

for col in cols_to_log:
    mpg_df[col + '_log'] = np.log1p(mpg_df[col])

mpg_df
```
