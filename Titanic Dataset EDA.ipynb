## Handle Outliers


```python
ti.describe()
```

```python
#outliers_columns

numerical_cols= ti.select_dtypes(include=[ 'int64','float64']).columns

outliers_columns={}
for col in numerical_cols:
    Q1= ti[col].quantile(0.25)
    Q3=ti[col].quantile(0.75)
    IQR=Q3-Q1

    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR

    outliers=ti[(ti[col]<lower_limit) | (ti[col]>upper_limit)]
    if len(outliers)>0:
        outliers_columns[col]=len(outliers)

print("Outliers_columns: ",outliers_columns)
```

```python
#IQR Method → (Q1, Q3, IQR, Upper, Lower bounds)
print("\nOutliers_Fare")

print("\nFirst Quartile: ",ti['fare'].quantile(0.25))
print("Second Quartile: ", ti['fare'].quantile(0.50))
print("Third Quartile: ",ti['fare'].quantile(0.75))
print("Minimum fare: ", ti['fare'].min())
print("Maximum fare: ", ti['fare'].max())
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
print("INTERQUARTILERANGE: ",Q3-Q1)
print("lower_limit: ", lower)
print("upper_limit: ", upper)
outliers = ti[(ti['fare'] < lower) | (ti['fare'] > upper)]
outliers.head()
```

```python
Q1=ti['age'].quantile(0.25)
Q2=ti['age'].quantile(0.50)
Q3=ti['age'].quantile(0.75)
print("\nOutliers_Age")
print("\nMinimum: ",ti['age'].min())
print("first quartile: ",Q1)
print("second quartile: ",Q2)
print("third quartile: ",Q3)
print("maximum: ",ti['age'].max())
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
print("lower bound: ",lower)
print("upper bound: ",upper)

outliers_age_df= ti[(ti['age']<lower) |  (ti['age']>upper)]
outliers_age_df
```

```python
Q1 = ti['sibsp'].quantile(0.25)
Q2 = ti['sibsp'].quantile(0.50)
Q3 = ti['sibsp'].quantile(0.75)
print("Outliers_Sibsp")
print("\nminimum: ", ti['sibsp'].min())
print("first quartile: ", Q1)
print("second quartile: ", Q2)
print("third quartile: ", Q3)
print("maximum: ", ti['sibsp'].max())

# IQR
IQR = Q3 - Q1

# Bounds
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("lower bound: ", lower)
print("upper bound: ", upper)

# Find outliers in sibsp
outliers_sibsp_df = ti[(ti['sibsp'] < lower) | (ti['sibsp'] > upper)]
outliers_sibsp_df
```

```python
Q1 = ti['parch'].quantile(0.25)
Q2 = ti['parch'].quantile(0.50)
Q3 = ti['parch'].quantile(0.75)

print("minimum: ", ti['parch'].min())
print("first quartile: ", Q1)
print("second quartile: ", Q2)
print("third quartile: ", Q3)
print("maximum: ", ti['parch'].max())

# IQR
IQR = Q3 - Q1

# Bounds
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("lower bound: ", lower)
print("upper bound: ", upper)

# Find outliers in parch
outliers_parch_df = ti[(ti['parch'] < lower) | (ti['parch'] > upper)]
outliers_parch_df
```

# Univariate Analysis (Single column analysis)

### Numerical → Histogram


```python
numerical_cols= ti.select_dtypes(include=[ 'int64','float64']).columns
numerical_cols
```

```python
num_cols = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']

for col in num_cols:
    plt.figure(figsize=(10,5))
    cat_hist=sns.histplot(data=ti, x=col, bins=30)
    patches= cat_hist.patches
    for patch in patches:
        patch.set_facecolor(np.random.rand(3,))


plt.title(f"Histplot of {col} distribution")
plt.xlabel(f"{col}")
plt.ylabel("Frequency")

plt.show()
```

## Numerical → KDE plot


```python
kde_cols = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']



plt.figure(figsize=(12, 8))

for col in kde_cols:
    sns.kdeplot(ti[col].dropna(), fill=True,alpha=0.3,linewidth=2, label=col)

    plt.title(f"KDE Plot of {col}")
    plt.xlabel(f"{col}")
    plt.ylabel("Density")
    plt.legend()
    plt.show()
```

### Categorical → Countplot


```python
# Countplot
categorical_cols=ti.select_dtypes("object").columns
for col in categorical_cols:
    plt.figure(figsize=(10,5))
    sns.countplot(x=col, hue=col, data=ti, palette='Set2', legend=False)
    plt.title(f'Countplot of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45) 
    plt.show()
```

# Categorical vs Numerical → Boxplot


```python
numerical_cols= ti.select_dtypes(include=[ 'int64','float64']).columns
numerical_cols
```

```python
categorical_cols= ti.select_dtypes(include=[ 'object']).columns
categorical_cols
```

```python
categorical_cols = ['sex', 'embarked', 'who', 'embark_town', 'alive']
numerical_cols = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']

for cat in categorical_cols:
    for num in numerical_cols:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=cat, y=num, data=ti)
        plt.title(f'Boxplot of {num} by {cat}')
        plt.xlabel(cat)
        plt.ylabel(num)
        plt.grid(True)
        plt.show()
```

# Numerical vs Numerical → Scatterplot / Correlation
